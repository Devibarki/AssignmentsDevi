```python
tweet=pd.read_csv(r'Elon_musk.csv',encoding='Latin-1')
tweet
```

# Preprocessing

```python
tweet.drop(['Unnamed: 0'],inplace=True,axis=1)
tweet
```

```python
tweet = [Text.strip() for Text in tweet.Text] # remove both the leading and the trailing characters
tweet = [Text for Text in tweet if Text] # removes empty strings, because they are considered in pyhton as false
tweet[0:10]
```

```python
tweet_text=' '.join(tweet)
tweet_text
```

```python
no_punc_text = tweet_text.translate(str.maketrans('', '', string.punctuation))
no_punc_text
```

# Tokenization

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
print(text_tokens[0:50])
```

```python
len(text_tokens)
```

# Remove StopWords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
my_stop_words = stopwords.words('english')
my_stop_words.append('the')
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:40])
```

# Normalize the data

```python
lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:25])
```

Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:40])
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(' '.join(no_stop_tokens))
print(doc[0:40])
```

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas[0:25])
```

Feature Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(lemmas)
```

```python
print(vectorizer.vocabulary_)
```

```python
print(vectorizer.get_feature_names()[50:100])
print(X.toarray()[50:100])
```

```python
print(X.toarray().shape)
```

Including Trigrams and Bigrams

```python
vectorizer_ngram_range = CountVectorizer(analyzer='word',ngram_range=(1,3),max_features = 100)
bow_matrix_ngram =vectorizer_ngram_range.fit_transform(tweet)
```

```python
print(vectorizer_ngram_range.get_feature_names())
print(bow_matrix_ngram.toarray())
```

TFidf vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer_n_gram_max_features = TfidfVectorizer(norm="l2",analyzer='word', ngram_range=(1,3), max_features = 500)
tf_idf_matrix_n_gram_max_features =vectorizer_n_gram_max_features.fit_transform(tweet)
print(vectorizer_n_gram_max_features.get_feature_names())
print(tf_idf_matrix_n_gram_max_features.toarray())
```

Generate WorldCloud

```python
def plot_cloud(wordcloud):
        plt.figure(figsize=(40, 30))
        plt.imshow(wordcloud) 
        plt.axis("off");
```

```python
stopwords = STOPWORDS
stopwords.add('will')
wordcloud = WordCloud(width = 3000, height = 2000, background_color='black', max_words=100,colormap='Set2',stopwords=stopwords).generate(tweet_text)
plot_cloud(wordcloud)
```
